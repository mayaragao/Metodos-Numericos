#### Método da Bisseção:

```python
tol = 0.0001
```

```python
def ordena(a, b):
    if a < b:
        return a, b
    return b, a
```

```python
def bissecao(a, b, tol):
    a, b = ordena(a, b)
    err = abs(b-a)
    aux = a
    it = 0

    while err >= tol:
        x = (a+b)/2.0
        print("ITER ", it, "|a = %.5f" % a, "|b = %.5f" % b, "|x = %.5f" % x,
              "|f(x) = %.4f" % f(x), "| Err = %.4f" % err)
        if f(x)*f(a) < 0:
            b = x
        elif f(x)*f(a) > 0:
            a = x
        else:
            return x
        err = abs(x-aux)
        aux = x
        it += 1
    print("ITER ", it, "|a = %.5f" % a, "|b = %.5f" % b, "|x = %.5f" % x,
          "|f(x) = %.4f" % f(x), "| Err = %.4f" % err)
    return x
```

#### Método de Newton original:

Definindo a derivada de $f(x)$

```python
def fder(x):
    cte = (9.806*0.00341)**0.5
    return (cte*np.sinh(x * cte)/np.cosh(x * cte))
```

```python
def newton(x0, tol, it_max):
    it = 0
    err = 10

    while (err >= tol) and (it < it_max):
        x = x0 - f(x0)/fder(x0)
        err = abs(x-x0)
        print("ITER ", it, "|x = %.3f" % x0, "|f(x) = %.4f" %
              f(x0), "| f'(x) = %.4f" % fder(x0), "| Err = %.5f" % err)
        x0 = x
        it += 1

    return (x, it)
```

```python
x, i = newton(x0, tol, it_max)
if i == it_max:
    print("O método não convergiu")
print("\nRaiz encontrada x= %.5f" % x)
```

#### Método da Secante:

```python
def secante(x0, tol, it_max):
    delta = 0.001
    x1 = x0 + delta
    fa = f(x0)
    it = 0
    err = 10
    while (err >= tol) and (it < it_max):
        fi = f(x1)
        x2 = x1 - (fi*(x1-x0)/(fi-fa))
        err = abs(x2-x1)
        print("ITER ", it, "|x", it, " = %.3f" % x0, "|x", it+1, " = %.3f" % x1,
              "|x", it+2, " = %.4f" %x2, "| Err = %.5f" % err)
        x0 = x1
        x1 = x2

        fa = fi
        it += 1

    return (x1, it)
```

```python
x0 = 10
x, i = secante(x0, tol, it_max)
if i == it_max:
    print("O método não convergiu")
print("\nRaiz encontrada x= %.5f" % x)
```

#### Método da Interpolação inversa:

```python
x = [200, 250, 300]
```

```python
def maior_indice(y):
    if y[0] > y[1]:
        maior = 0
    else:
        maior = 1
    if maior > y[2]:
        return maior
    return 2
```

```python
def interpolacao_inversa(x):
    it = 0
    err = 10
    x = sorted(x)
    x0 = 10**10
    y = [0, 0, 0]
    y[0] = f(x[0])
    y[1] = f(x[1])
    y[2] = f(x[2])

    while (err >= tol) and (it < it_max):
        phi0 = ((y[1]*y[2])/((y[0]-y[1])*(y[0]-y[2])))
        phi1 = ((y[0]*y[2])/((y[1]-y[0])*(y[1]-y[2])))
        phi2 = ((y[0]*y[1])/((y[2]-y[0])*(y[2]-y[1])))
        aux = phi0*x[0] + phi1*x[1] + phi2*x[2]

        err = abs(aux - x0)

        print("ITER ", it, "|x", it, " = %.3f" % x[0],
              "|x", it+1, " = %.3f" % x[1], "|x", it+2, " = %.3f" %
              x[2], "|y", it, " = %.3f" % y[0],
              "|y", it+1, " = %.3f" % y[1], "|y", it+2, " = %.4f" %
              y[2], "| x* = %.3f" % aux, "\n        Err = %.5f" % err)

        i = maior_indice(y)
        x[i] = aux
        y[i] = f(aux)
        x = sorted(x)
        y = sorted(y)
        x0 = aux
        it += 1

    return (aux, it)
```

```python
aux, i = interpolacao_inversa(x)
if i == it_max:
    print("O método não convergiu")
print("\nRaiz encontrada x= %.5f" % aux)
```

### Exercício 3)

$$16x^4 + 16y^4 + z^4 = 16$$

$$x^2 + y^2 + z^2 = 3$$

$$x^3 -y + z = 1$$

#### Método de Newton:

```python
def f(V):
    x, y, z = V
    S = np.array([16*x**4+16*y**4+z**4-16,
                  x**2+y**2+z**2-3,
                  x**3-y+z-1
                  ])
    return S
```

```python
def j(V):
    x, y, z = V
    J = np.array([[64*x**3, 64*y**3, 4*z**3],
                  [2*x, 2*y, 2*z],
                  [3*x**2, -1, 1]
                  ])
    return J
```

```python
def newton(V, tol, it_max):
    it = 0
    err = 10

    while (err >= tol) and (it < it_max):
        J = j(V)
        F = f(V)
        inv_J = np.linalg.inv(J)

        delta = - inv_J.dot(F)
        V = V + delta
        err = np.linalg.norm(delta)/np.linalg.norm(V)

        X = np.round(V, 3)
        print("ITER ", it, "|X = ", X, " | Err = %.5f" % err)

        it += 1

    return (V, it)
```

Definindo as condições iniciais

```python
V = np.array([1, 1, 1])
tol = 0.0001
it_max = 100
```

```python
X, i = newton(V, tol, it_max)
X = np.round(X, 3)
if i == it_max:
    print("O método não convergiu")
print("\nSolução X = ",  X)
```

#### Método de Broyden:

```python
def broyden(X, B):
    it = 0
    err = 10
    while (err >= tol) and (it < it_max):
        J = np.copy(B)

        F = f(X)
        F = np.reshape(F, (-1, 1))

        delta = np.linalg.solve(J, -F)

        X = np.reshape(X, (-1, 1)) + delta
        X = X + delta

        Y = f(X)
        Y = np.reshape(Y, (-1, 1))
        Y = Y - F
        err = np.linalg.norm(delta)/np.linalg.norm(X)

        print("ITER ", it, "|X = ", np.reshape(np.round(X, 3), (1, -1)), " | Err = %.5f" % err)

        aux = (Y - np.dot(B, delta))
        numerador = np.dot(aux, np.transpose(delta))
        denominador = np.dot(np.transpose(delta), delta)
        B = B + np.divide(numerador, denominador)
        it += 1

    return (X, it)
```

Para matriz $B$ inicial sendo o jacobiano do vetor $X$:

```python
X = np.array([1, 1, 1])
B = j(X)
```

```python
print(B)
```

```python
X, i = broyden(X, B)
if i == it_max:
    print("O método não convergiu")
print("\nSolução X = ",  np.reshape(np.round(X, 3), (1, -1)))
```

Para matriz $B$ identidade:

```python
X = np.array([1, 1, 1])
B = np.eye(3)
```

```python
X, i = broyden(X, B)
if i == it_max:
    print("O método não convergiu")
print("\nSolução X = ",  np.reshape(np.round(X, 3), (1, -1)))
```

Percebemos que são soluções diferentes, contudo, ambas são soluções do sistema de equações definido.

### Exercício 5)

$$f(x) = b_0 +b_1 x^{b_2} $$

$$x = [1,2,3] $$
$$y = [1,2,9] $$

#### Mínimos quadrados

Rotina para definir a funcao $f(x)$ e definir o jacobiano $j(x)$ de $f$ .

```python
def f(B):
    X = np.array([1, 2, 3])
    Y = np.array([1, 2, 9])
    b0, b1, b2 = B

    S = np.array([b0 + b1*(X[0]**b2) - Y[0],
                  b0 + b1*(X[1]**b2) - Y[1],
                  b0 + b1*(X[2]**b2) - Y[2]
                  ])
    return S
```

```python
def j(B):
    b0, b1, b2 = B

    J = np.array([[1,  1,  0],
                  [1,  2**b2,  b1*(2**b2)*np.log(2)],
                  [1,  3**b2,  b1*(3**b2)*np.log(3)]
                  ])
    return J
```

```python
def minimos(B, tol, it_max):
    it = 0
    err = 10

    while (err >= tol) and (it < it_max):
        J = j(B)
        F = f(B)
        delta = np.linalg.solve(
            np.dot(np.transpose(J), J),  np.dot(np.transpose(J), -F))
        B = B + delta
        err = np.linalg.norm(delta)/np.linalg.norm(B)
        x = np.round(B, 3)
        print("ITER ", it, "|X = ",  x, " | Err = %.5f" % err)
        it += 1

    return (B, it)
```

Definindo valores iniciais:

```python
B = np.array([0, 1, 2])
tol = 0.0001
it_max = 100
```

```python
V, i = minimos(B, tol, it_max)
V = np.round(V, 4)
if i == it_max:
    print("O método não convergiu")
print("\nSolução X = ",  V)
```
